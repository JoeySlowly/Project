```python
tnx = yf.download('^TNX', start='2020-01-01', end='2023-12-31')
tnx
```

```python
crude_oil = yf.download('CL=F', start='2020-01-01', end='2023-12-31')
crude_oil 
```

```python
xle_returns = returns[['XLE']].copy()
xle_returns.rename(columns={'XLE': 'XLE_Returns'}, inplace=True)
xle_returns
```

```python
tnx_cleaned = tnx[['Adj Close']].copy()
tnx_cleaned.rename(columns={'Adj Close': 'Interest_Rate'}, inplace=True)
crude_cleaned = crude_oil[['Adj Close']].copy()
crude_cleaned.rename(columns={'Adj Close': 'Crude_Oil_Price'}, inplace=True)
```

```python
merged_data = pd.concat([xle_returns, tnx_cleaned, crude_cleaned], axis=1)
merged_data
```

```python
merged_data.dropna(inplace=True)
merged_data.rename(columns={
    'XLE_Returns': 'XLE_Returns',
    '(Interest_Rate, ^TNX)': 'Interest_Rate',
    '(Crude_Oil_Price, CL=F)': 'Crude_Oil_Price'
}, inplace=True)
merged_data
```

```python
import statsmodels.api as sm
```

```python
merged_data.rename(columns={
    '(Interest_Rate, ^TNX)': 'Interest_Rate',
    '(Crude_Oil_Price, CL=F)': 'Crude_Oil_Price'
}, inplace=True)
```

```python
merged_data.columns = ['XLE_Returns', 'Interest_Rate', 'Crude_Oil_Price']
```

```python
print(merged_data.columns)
```

```python
Y = merged_data['XLE_Returns']
X = merged_data[['Interest_Rate', 'Crude_Oil_Price']]
X = sm.add_constant(X) 
model = sm.OLS(Y, X).fit()
print(model.summary())
```

P > 0.05

```python
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
```

```python
def check_stationarity(series, name):
    result = adfuller(series.dropna())
    print(f'{name} ADF Statistic: {result[0]}')
    print(f'{name} p-value: {result[1]}')
    if result[1] > 0.05:
        print(f"{name} is not stationary. Differencing may be needed.")
    else:
        print(f"{name} is stationary.")

check_stationarity(merged_data['XLE_Returns'], 'XLE_Returns')
check_stationarity(merged_data['Interest_Rate'], 'Interest_Rate')
check_stationarity(merged_data['Crude_Oil_Price'], 'Crude_Oil_Price')
```

```python
merged_data['Interest_Rate_Diff'] = merged_data['Interest_Rate'].diff()
merged_data['Crude_Oil_Price_Diff'] = merged_data['Crude_Oil_Price'].diff()

check_stationarity(merged_data['Interest_Rate_Diff'], 'Interest_Rate_Diff')
check_stationarity(merged_data['Crude_Oil_Price_Diff'], 'Crude_Oil_Price_Diff')
```

```python
from statsmodels.tsa.api import VAR
var_data = merged_data[['XLE_Returns', 'Interest_Rate_Diff', 'Crude_Oil_Price_Diff']].dropna()
var_model = VAR(var_data)
var_results = var_model.fit(maxlags=6)  
print(var_results.summary())
```

It's working better!

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
lb_test = acorr_ljungbox(var_results.resid['XLE_Returns'], lags=[10], return_df=True)
print(lb_test)
```
